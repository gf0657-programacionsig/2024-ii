## Introducción

[matplotlib](https://matplotlib.org/) es una de las bibliotecas más populares de Python para visualización de datos. Se utiliza principalmente para crear gráficos estáticos en dos dimensiones, pero también puede emplearse para generar visualizaciones animadas, interactivas y en tres dimensiones.

matplotlib se compone de varios módulos, siendo [`pyplot`](https://matplotlib.org/stable/api/pyplot_summary.html) el principal y el más utilizado.

## Instalación y carga

```python
# Carga del módulo pyplot de matplotlib
import matplotlib.pyplot as plt
```

## Interfaces

matplotlib cuenta con dos [interfaces](https://matplotlib.org/stable/api/index) para crear visualizaciones: una basada en la clase `Axes` y otra basada en el módulo `pyplot`.

### Interfaz basada en la clase `Axes`

Esta interfaz es orientada a objetos y utiliza métodos para tareas como agregar datos, configurar los límites de los ejes y especificar etiquetas. Permite controlar una gran cantidad de detalles y se recomienda para visualizaciones complejas o con múltiples subgráficos.

La estructura principal de una visualización programada con esta interfaz consta de objetos de dos clases:

- [`Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html) (figura): es el contenedor general del gráfico. Puede verse como un canvas o lienzo en el que se dibujan los gráficos.
- [`Axes`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.html) (subgráficos): son los "subgráficos" o partes de la figura en los que se dibujan los datos. Un objeto de la clase `Axes` contiene elementos visuales como geometrías (líneas, barras, histogramas, etc.), títulos y etiquetas.

En el siguiente bloque de código se genera una figura con un subgráfico.

```python
# Figura con:
# - Gráfico de dispersión que muestra la relación entre estatura y peso.

# Datos
estatura = [1.77, 1.57, 1.8, 1.6, 1.63, 1.69, 1.75, 1.8, 1.9, 1.55, 1.65, 1.71, 1.85, 1.73, 1.62]
peso = [75.1, 62.9, 78.4, 61.2, 67.8, 71.1, 75.5, 81.6, 82.2, 64.4, 65.6, 72.2, 90.3, 74.3, 66.7]

# Creación una figura con un subgráfico
fig, ax = plt.subplots()

# Creación de un gráfico de dispersión
ax.scatter(estatura, peso)

# Personalización del gráfico
ax.set_title('Relación entre estatura y peso')
ax.set_xlabel('Estatura (m)')
ax.set_ylabel('Peso (kg)')

# Despliegue de la figura
plt.show()
```

En este ejemplo se utilizaron los siguientes métodos:

1. El método [`matplotlib.pyplot.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) retorna, en una tupla, una figura y un conjunto de subgráficos. La figura se almacena en el objeto `fig` y el único subgráfico en el objeto `ax`. 
2. El método [`matplotlib.axes.Axes.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html) genera un gráfico de dispersión (*scatter plot*) en el subgráfico `ax`.
3. Los métodos [`matplotlib.axes.Axes.set_title()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html), [`matplotlib.axes.Axes.set_xlabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html) y [`matplotlib.axes.Axes.set_ylabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html) ayudan a personalizar el subgráfico `ax` al especificar respectivamente el título, la etiqueta del eje x y la etiqueta del eje y. 
4. Por último, el método [`matplotlib.pyplot.show()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html) despliega todas las figuras que han sido creadas. En este caso, solo se despliega `fig`. Una figura específica puede desplegarse con el método [`matplotlib.figure.Figure.show()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.show.html). Sin embargo, debe tenerse en cuenta que este último método puede no funcionar adecuadamente en ambientes interactivos como cuadernos de notas de Jupyter.

La interfaz basada en subgráficos también puede utilizarse para generar figuras con varios subgráficos, como en el siguiente bloque de código:

```python
# Figura con:
# - Gráfico de dispersión que muestra la relación entre estatura y peso.
# - Gráfico de dispersión que muestra la relación entre peso y grasa corporal.

# Datos
estatura = [1.77, 1.57, 1.8, 1.6, 1.63, 1.69, 1.75, 1.8, 1.9, 1.55, 1.65, 1.71, 1.85, 1.73, 1.62]
peso = [75.1, 62.9, 78.4, 61.2, 67.8, 71.1, 75.5, 81.6, 82.2, 64.4, 65.6, 72.2, 90.3, 74.3, 66.7]
grasa_corporal = [19.43, 20.88, 19.59, 19.29, 20.96, 20.53, 20.09, 20.7, 17.96, 22.67, 19.35, 20.25, 22.26, 20.37, 20.88]

# Creación de una figura de 8x6 pulgadas
# con dos subgráficos distribuidos en 2 filas y 1 columna
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Creación del primer gráfico de dispersión en ax1
ax1.scatter(estatura, peso, c='blue')

# Creación del segundo gráfico de dispersión en ax12
ax2.scatter(peso, grasa_corporal, c='red')

# Personalización de ax1
ax1.set_title('Relación entre estatura y peso')
ax1.set_xlabel('Estatura (m)')
ax1.set_ylabel('Peso (kg)')

# Personalización de ax2
ax2.set_title('Relación entre peso y grasa corporal')
ax2.set_xlabel('Peso (kg)')
ax2.set_ylabel('Grasa corporal (%)')

# Ajuste del espacio entre los subgráficos
plt.tight_layout()

# Despliegue de la figura
plt.show()
```

En este caso, el método `matplotlib.pyplot.subplots()` fue llamado con los argumentos `nrows` y `ncols` correspondientes a la cantidad de filas (1) y de columnas (2), lo que proporciona espacio para dos subgráficos: `ax1` y `ax2`. El argumento `figsize` especifica el tamaño de la figura en pulgadas (8 de ancho x 6 de alto). El método [`matplotlib.pyplot.tight_layout()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html) ajusta el espacio entre y alrededor de los subgráficos. En `matplotlib.axes.Axes.scatter()` se usa el argumento `c` para especificar el color de los puntos.

### Interfaz basada en el módulo `pyplot`

Esta interfaz es conveniente para gráficos simples y rápidos. La mayoría de las funciones se llaman directamente desde `pyplot`. Las clases `Figure` y `Axes` se manipulan a través de estas funciones, sin tener que declarar objetos, por lo que se dice que se usan de manera "implícita".

El siguiente bloque de código genera un gráfico mediante esta interfaz.

```python
# Figura con:
# - Gráfico de dispersión que muestra la relación entre estatura y peso.

# Datos
estatura = [1.77, 1.57, 1.8, 1.6, 1.63, 1.69, 1.75, 1.8, 1.9, 1.55, 1.65, 1.71, 1.85, 1.73, 1.62]
peso = [75.1, 62.9, 78.4, 61.2, 67.8, 71.1, 75.5, 81.6, 82.2, 64.4, 65.6, 72.2, 90.3, 74.3, 66.7]

# Creación de un gráfico de dispersión
plt.scatter(estatura, peso)

# Personalización del gráfico
plt.title('Relación entre estatura y peso')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')

# Despliegue de la figura
plt.show()
```

En esta visualización, las funciones [`matplotlib.pyplot.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html), [`matplotlib.pyplot.title()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html), [`matplotlib.pyplot.xlabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) y [`matplotlib.pyplot.ylabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html) y `matplotlib.pyplot.show()` acceden a una figura y a un gráfico que han sido creados de manera implícita por `pyplot`.

## Integración con pandas

pandas cuenta con un sistema de graficación basado en matplotlib, lo que permite crear gráficos directamente a partir de series y dataframes de pandas.



### Tipos de gráficos

El método [`pandas.DataFrame.plot()`](pandas.DataFrame.plot) genera gráficos de diferentes tipos, los cuales se especifican por medio del argumento `kind` (kind='bar', kind='pie', kind='hist', kind='scatter', etc.).

Cada tipo de gráfico puede generarse también a través de funciones como `pandas.DataFrame.plot.bar()`, `pandas.DataFrame.plot.pie()`, `pandas.DataFrame.plot.hist()`, `pandas.DataFrame.plot.scatter()` y otras.

```python
import pandas as pd

# Configuración de pandas para mostrar separadores de miles y 2 dígitos decimales
pd.set_option('display.float_format', '{:,.2f}'.format)

# Carga de datos de países en un dataframe
paises = pd.read_csv(
    "https://raw.githubusercontent.com/gf0657-programacionsig/2024-ii/refs/heads/main/datos/natural-earth/paises.csv"
)

# Se usa la columna ADM0_ISO como índice
paises.set_index('ADM0_ISO', inplace=True)

# Despliegue de una muestra aleatoria de 5 filas
paises.sample(5)
```

#### Gráficos de dispersión

```python
from matplotlib import ticker

# Subconjunto de países de América Central
paises_america_central = paises.loc[['PAN', 'CRI', 'NIC', 'SLV', 'HND', 'GTM', 'BLZ']]

# Creción de una figura y un subgráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Creación del gráfico de dispersión
paises_america_central.plot.scatter(
    ax=ax, 
    x='POP_EST', 
    y='GDP_MD', 
    color='red',
    s=50
)

# Personalización del gráfico
ax.set_title('Relación entre población y producto interno bruto (PIB) en países de América Central')
ax.set_xlabel('Población (habitantes)')
ax.set_ylabel('PIB (millones de dólares)')

# Configurar separadores de miles en los ejes y evitar notación científica
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Despliegue del gráfico
plt.show()
```
